```python
from urllib.request import urlopen
import numpy as np
import pandas as pd
import requests
import json
```

```python
query_str = 'இலங்கை+மலையகம்'
myResponse = requests.get("https://www.googleapis.com/books/v1/volumes?q="+ query_str +"&key=GoogleAPIKey&startIndex=1&maxResults=40")
jData = json.loads(myResponse.content)
print(jData)
```

```python

def get_book_info(query_str, startIndex, df):
    book_info = requests.get("https://www.googleapis.com/books/v1/volumes?q="+ query_str +"&key=GoogleAPIKey&startIndex="+ str(startIndex) +"&maxResults=40")
    
    json_data = json.loads(book_info.content)
    items = json_data["items"]
    
    for item in items:
        local_id = item["id"]
        volumeInfo = item["volumeInfo"]
        title = volumeInfo.get("title")
        author = volumeInfo.get("authors")
        description = volumeInfo.get("description")
        publisher = volumeInfo.get("publisher")
        datePublished = volumeInfo.get("publishedDate")
        extent = volumeInfo.get("pageCount")
        printType = volumeInfo.get("printType")
        keywords = volumeInfo.get("categories")
        language = volumeInfo.get("language")
        canonicalVolumeLink = volumeInfo.get("canonicalVolumeLink")
        industry_id = ""
        if 'industryIdentifiers' in volumeInfo:
            industry_id_node = volumeInfo["industryIdentifiers"][0]                  
            industry_id = industry_id_node.get("identifier")
        df = df.append({'id':local_id, 'industry_id': industry_id, 'title':title, 'author': author, 'keywords':keywords, 'publisher':publisher, 'datePublished':datePublished, 'description':description,'extent': extent, 'language': language, 'type': printType, 'link': canonicalVolumeLink}, ignore_index=True)    
        
    return df
    
```

```python
total_items = jData["totalItems"]
total_items = int(total_items)
startIndex = 0
max_count = int(total_items / 40)
count = 0
df = pd.DataFrame()
while count <= max_count:
    df = get_book_info(query_str, count, df)
    count = count + 1   
```

```python
df = df[['id', 'title', 'author', 'keywords', 'publisher', 'datePublished', 'type', 'description', 'extent', 'language', 'industry_id', 'link']]
df.to_csv("google9.csv")
```
