```python
import numpy as np
import pandas as pd
```

```python
collection1 = pd.read_csv("books.csv");
collection1["தலைப்பு"] = collection1["தலைப்பு"].str.strip()
collection1["collection1_title"] = collection1["தலைப்பு"].str.replace(" ","")
collection1.head()
```

```python
def split_test(row):
    title = row["title"]
    new_title = title.split("/", 1)[0]
    return new_title

collection2 = pd.read_csv("books_opac.csv");
collection2["title"] = collection2["title"].str.strip()
collection2["new_title"] = collection2.apply(split_test, axis=1)
collection2["collection2_title"] = collection2["new_title"].str.replace(" ","")
collection2.head()
```

```python
result = collection1.merge(collection2, left_on='collection1_title', right_on='collection2_title')
result.to_csv("test2.csv")
```
