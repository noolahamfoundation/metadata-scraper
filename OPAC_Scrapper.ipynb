```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
def get_work_info(work_url, df):
    page = "http://opacip:8083" + work_url
    page = urlopen(page)
    soup = BeautifulSoup(page, "html.parser")

    title = soup.find("h1", attrs={"class": "title"}).text.strip()

    author = ""
    author_nodes = soup.findAll("span", attrs={"property": "author"})
    if author_nodes is not None:
        for row in author_nodes:
            author = author + " | " + row.text
        author =  author[2:]
         
    contributor = ""
    contributor_nodes = soup.findAll("span", attrs={"property": "contributor"})
    if contributor_nodes is not None:
        for row in contributor_nodes:
            contributor = contributor + " | " + row.text
        contributor =  contributor[2:]
            
    keywords = ""
    keywords_node = soup.find("span", attrs={"class": "subjects"})
    if keywords_node is not None:
        keywords = keywords_node.text.replace("Subject(s):", "").strip()
            
    publisher = ""
    publisher_node = soup.find("span", attrs={"property": "name"})
    if publisher_node is not None:
        publisher = publisher_node.text.replace(",", "")
    
    datePublished = ""
    datePublished_node = soup.find("span", attrs={"property": "datePublished"})
    if datePublished_node is not None:
        datePublished = datePublished_node.text
            
    location = ""
    location_node = soup.find("span", attrs={"property": "location"})
    if location_node is not None:
        location = location_node.text.replace(":", "")
        
    description = ""
    description_node = soup.find("span", attrs={"property": "description"})
    if description_node is not None:
        description = description_node.text

    work_type = ""
    work_type_node = soup.find("span", attrs={"class": "type"})
    if work_type_node is not None:
        work_type = work_type_node.text
 
    ddc = ""
    ddc_node = soup.find("span", attrs={"class": "ddc"})
    if ddc_node is not None:
        ddc = ddc_node.text.replace("DDC classification:", "").strip()

    print(title)
    item_df = df.append({'id':work_url, 'title': title, 'author':author, 'contributor':contributor, 'keywords':keywords, 'publisher':publisher, 'datePublished':datePublished, 'location': location, 'description':description, 'type':work_type, 'ddc': ddc}, ignore_index=True)
    return item_df
```

```python
def get_page_info(offset, df):
    page = "http://opacip:8083/cgi-bin/koha/opac-search.pl?idx=kw&q=tam&offset="+ offset +"&sort_by=relevance_dsc&limit=itype:BOOK"
    page = urlopen(page)
    soup = BeautifulSoup(page, "html.parser")
    works = soup.findAll("a", attrs={"class": "p1"})

    for work in works:
        item_attrs = work.attrs
        work_info = item_attrs['href']
        print (work_info)
        df = get_work_info(work_info, df)
    return df
```

```python
df = pd.DataFrame()

count = 0
while count < 81:
    df = get_page_info(str(count), df)
    count = count + 20
    
df 
```

```python
df = df[['title', 'author', 'contributor', 'keywords', 'publisher', 'datePublished', 'location', 'type', 'description', 'ddc', 'id']]
df.to_csv("books_opac.csv")
df.head()
```
